```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Heat Map


```python
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscore = tips[['total_bill', 'tip','size']]
tipscore    
```

```python
sns.heatmap(tipscore.corr(), annot=True, cmap='coolwarm')
```

Cluster Map


```python
sns.clustermap(tipscore.corr(), annot=True)
```

pivot table Heat map

```python
pvtflight = flights.pivot_table(index='month', columns='year', values='passengers')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
